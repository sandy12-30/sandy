```python
import cv2
import os
import glob
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
```

##  元ファイルを200x200に縮小する

```python
darray=["00Blue","01Green","02Noukin","03Hakukin"]

for dname in darray:
    dirname = 'CAPDB_pub/'+dname+'/ROI/*/org_*.bmp'
    files = glob.glob(dirname)
    for f in files:
        img = Image.open(f)
        img_resize = img.resize((200, 200))
        ftitle, fext = os.path.splitext(f)
        fname= ftitle.replace("org_","sqr_")+fext
        print(fname)
        img_resize.save(fname)
```

## Sobelフィルタの生成と変換

```python
for dname in darray:
    dirname = 'CAPDB_pub/'+dname+'/ROI/*/sqr_*.bmp'
    files = glob.glob(dirname)
    for f in files:
        print(f)
        img = cv2.imread(f)
        #Gray Scaleへ変換
        
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        
        gray_sobelx = cv2.Sobel(gray,cv2.CV_32F,1,0)#x方向のエッジ検出
        gray_sobely = cv2.Sobel(gray,cv2.CV_32F,0,1)#y方向のエッジ検出

   
        #8ビット符号なし整数変換
        gray_abs_sobelx = cv2.convertScaleAbs(gray_sobelx) 
        gray_abs_sobely = cv2.convertScaleAbs(gray_sobely)
        
        #重み付き和
        gray_sobel_edge = cv2.addWeighted(gray_abs_sobelx,0.5,gray_abs_sobely,0.5,0) 
        fname = f.replace("sqr_","sbf_")
        print(fname)
        cv2.imwrite(fname, gray_sobel_edge)

```
